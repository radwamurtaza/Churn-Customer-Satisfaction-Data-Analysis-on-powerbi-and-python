```python
# Handle categorical variables using one-hot encoding
df = pd.get_dummies(df, columns=[ 'satisfaction_v2'])
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(15,10))
plt.title('Correlation Matrix')
sns.heatmap(correlation_matrix, annot=True, linewidths=0.5, fmt=".2f", cmap="YlGnBu")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


# Remove leading/trailing whitespaces from column names
df.columns = df.columns.str.strip()

# Display the updated list of columns
print(df.columns)

    # Split the dataset into features (X) and target variable (y)
X = df.drop('satisfaction_v2_satisfied', axis=1)
y = df['satisfaction_v2_satisfied']
    # Define class labels for visualization
class_labels = ['Satisfied Customer', 'Unsatisfied Customer']
X_encoded = pd.get_dummies(X)

```

```python
 # Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Initialize the Random Forest Classifier
rfc = RandomForestClassifier(n_estimators=100, random_state=15)

# Fit the model on the training data
rfc.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rfc.predict(X_test)

# Calculate the accuracy of the model
rfc_accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', rfc_accuracy)

# Calculate the classification_report of the model
rfc_report = classification_report(y_test, y_pred)
print('Classification Report:\n', rfc_report)

# Calculate the confusion matrix of the model
rfc_cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:', rfc_cm)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(rfc_cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=class_labels, yticklabels=class_labels)

plt.title('Confusion Matrix - RandomForestClassifier')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)

reg_accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', reg_accuracy)

reg_report = classification_report(y_test, y_pred)
print('Classification Report:\n', reg_report)

reg_cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:', reg_cm)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(reg_cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=class_labels, yticklabels=class_labels)

plt.title('Confusion Matrix - LogisticRegression')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)

gnb_accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', gnb_accuracy)

gnb_report = classification_report(y_test, y_pred)
print('Classification Report:\n', gnb_report)

gnb_cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:', gnb_cm)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(gnb_cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=class_labels, yticklabels=class_labels)

plt.title('Confusion Matrix - GaussianNB')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)

dtc_accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', dtc_accuracy)

dtc_report = classification_report(y_test, y_pred)
print('Classification Report:\n', dtc_report)

dtc_cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:', dtc_cm)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(dtc_cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=class_labels, yticklabels=class_labels)

plt.title('Confusion Matrix - DecisionTreeClassifier')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
model_accuracy = {
    "Random Forest Classifier" : rfc_accuracy*100,
    "Logistic Regression" : reg_accuracy*100,

    "Gaussian Naive Bayes" : gnb_accuracy*100,
    "DecisionTreeClassifier" : dtc_accuracy*100
}
model_accuracy
```

```python
model_compare = pd.DataFrame(model_accuracy, index=["accuracy"])
fig, ax = plt.subplots(figsize=(15, 5))
ax.bar(model_compare.columns, model_compare.iloc[0], color="#87CBB9")
plt.xlabel("Model Names")
plt.ylabel("Score")
# plt.xticks(rotation=0)
plt.title("Model Comparison");
```
